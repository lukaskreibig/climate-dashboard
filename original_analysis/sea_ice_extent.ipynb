```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

# Ensure Year is numeric and sort the DataFrame by Year and DayOfYear
df_ice_clean["Year"] = pd.to_numeric(df_ice_clean["Year"], errors="coerce")
df_ice_clean = df_ice_clean.sort_values(["Year", "DayOfYear"])

# Get sorted unique years
unique_years = np.sort(df_ice_clean["Year"].unique())
n_years = len(unique_years)

# Create color gradient using a built-in colorscale (here "Jet")
# For each year, sample a color using the normalized position in the years list.
colors = [px.colors.sample_colorscale("Jet", i/(n_years-1))[0] for i in range(n_years)]

# Create the figure and add one trace per year. 
# We hide each trace's legend entry to avoid a crowded legend.
fig = go.Figure()

for year, color in zip(unique_years, colors):
    df_year = df_ice_clean[df_ice_clean["Year"] == year]
    fig.add_trace(go.Scatter(
        x=df_year["DayOfYear"],
        y=df_year["Extent"],
        mode="lines",
        name=str(year),
        line=dict(color=color, width=1),
        showlegend=False 
    ))

# Add a dummy trace to show a continuous colorbar representing the Year range.
dummy_trace = go.Scatter(
    x=[None],
    y=[None],
    mode="markers",
    marker=dict(
        colorscale="Jet",
        color=[unique_years.max(), unique_years.min()],
        showscale=True,
        colorbar=dict(
            title="Year",
            tickvals=[unique_years.max(), round(unique_years.mean()), unique_years.min()],            
        )
    ),
    hoverinfo="none",
    showlegend=False
)
fig.add_trace(dummy_trace)

# Update layout for clarity.
fig.update_layout(
    title="Interactive Seasonal Trends of Sea Ice Extent Across Years",
    xaxis_title="Day of Year",
    yaxis_title="Sea Ice Extent (e.g., million sq km)",
    template="plotly_white"
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Ensure Month is categorical with the proper order:
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
df_ice_clean["Month"] = pd.Categorical(df_ice_clean["Month"], categories=month_order, ordered=True)

# Create a "SeasonDate" column using a constant year (e.g., 2000) so that the x-axis can be labeled with month names.
# This column is used only for plotting; all calculations use the full daily values.
df_ice_clean["SeasonDate"] = pd.to_datetime(
    "2000 " + df_ice_clean["Month"].astype(str) + " " + df_ice_clean["Day"].astype(str),
    format="%Y %B %d"
)

# Exclude 2025 from envelope calculations (since it's incomplete)
df_env = df_ice_clean[df_ice_clean["Year"] != 2025]

# Group by SeasonDate (i.e. every day of the year) to compute statistics across all available daily data.
df_stats = df_env.groupby("SeasonDate")["Extent"].agg(
    min_val = "min",               
    mean_val = "mean",           
    q25 = lambda x: x.quantile(0.25), 
    q75 = lambda x: x.quantile(0.75)  
).reset_index()

# 2025, compute the daily average (by SeasonDate)
df_2025 = df_ice_clean[df_ice_clean["Year"] == 2025].groupby("SeasonDate")["Extent"].mean().reset_index()

fig = go.Figure()

# Mean line (aggregated from daily data)
fig.add_trace(go.Scatter(
    x = df_stats["SeasonDate"],
    y = df_stats["mean_val"],
    name = "Mean (daily data aggregated)",
    mode = "lines",
    line = dict(color = "black", width = 2)
))

# Upper IQR Boundary (75th percentile) with legend entry
fig.add_trace(go.Scatter(
    x = df_stats["SeasonDate"],
    y = df_stats["q75"],
    name = "IQR (25th–75th)",
    legendgroup = "IQR",
    showlegend = True,
    mode = "lines",
    line = dict(width = 0)
))

# Lower IQR Boundary (25th percentile) and fill to the upper boundary
fig.add_trace(go.Scatter(
    x = df_stats["SeasonDate"],
    y = df_stats["q25"],
    name = "IQR (25th–75th)",
    legendgroup = "IQR",
    showlegend = False,
    mode = "lines",
    fill = "tonexty", 
    line = dict(width = 0),
    fillcolor = "rgba(173,216,230,0.2)"
))

# Absolute Minimum as a red dashed line
fig.add_trace(go.Scatter(
    x = df_stats["SeasonDate"],
    y = df_stats["min_val"],
    name = "Absolute Min (daily data)",
    mode = "lines",
    line = dict(color = "red", dash = "dash", width = 2)
))

# 2025 (partial) daily average as an orange line
fig.add_trace(go.Scatter(
    x = df_2025["SeasonDate"],
    y = df_2025["Extent"],
    name = "2025 (partial)",
    mode = "lines",
    line = dict(color = "orange", width = 2)
))

fig.update_layout(
    title = "Seasonal Sea Ice Extent: Envelope from Daily Data (Excl. 2025) with 2025 (partial)",
    xaxis_title = "Month",
    yaxis_title = "Sea Ice Extent (e.g., million sq km)",
    template = "plotly_white"
)

fig.update_xaxes(
    dtick="M1",       
    tickformat="%B"    
)

fig.show()
```

```python
import pandas as pd

# Compute the multi-year average for each DayOfYear.
seasonal_mean = df_ice_clean.groupby("DayOfYear")["Extent"].mean().reset_index(name="Seasonal_Mean")

# Merge back to get a daily anomaly: Extent - Seasonal_Mean
df_merged = df_ice_clean.merge(seasonal_mean, on="DayOfYear", how="left")
df_merged["DailyAnomaly"] = df_merged["Extent"] - df_merged["Seasonal_Mean"]

# Average anomalies by year to get an annual anomaly
df_annual = df_merged.groupby("Year", as_index=False)["DailyAnomaly"].mean()
df_annual.rename(columns={"DailyAnomaly": "AnnualAnomaly"}, inplace=True)

fig = px.bar(
    df_annual,
    x="Year",
    y="AnnualAnomaly",
    color="AnnualAnomaly",
    color_continuous_scale=["red", "white", "blue"],
    range_color=[df_annual["AnnualAnomaly"].min(), df_annual["AnnualAnomaly"].max()],
    labels={
        "Year": "Year",
        "AnnualAnomaly": "Anomaly"
    },
    title="Annual Sea Ice Extent Anomalies",
    template="plotly_white",
)

fig.add_hline(y=0, line_dash="dash", line_color="black")

fig.update_layout(
    coloraxis_showscale=True,
    xaxis=dict(dtick=1)
)

fig.show()
```

### Compute Annual Sea Ice Mean
We group the cleaned sea ice dataset by "Year" and compute the mean "Extent".
The result is stored in `sea_ice_annual` with a column named "SeaIceMean".

```python
sea_ice_annual = df_ice_clean.groupby("Year")["Extent"].mean().reset_index(name="SeaIceMean")
sea_ice_annual.head()
```

### Merge Sea Ice Data with Temperature Data
We merge the temperature DataFrame (`temp_df`) with the annual sea ice data on "Year".
The merge type is "inner" so that only years present in both datasets are kept.

```python
merged_tempice = pd.merge(temp_df, sea_ice_annual, on="Year", how="inner")
merged_tempice.head()
```

### Load and Process Annual CO₂ Emissions Data
We load the annual CO₂ emissions data from CSV. Then, we compute a new column "GlobalCO2Mean"
which is the mean CO₂ emissions for each year. We use `transform("mean")` so that the computed mean
is broadcasted to every row for that year.

```python
annual_co_emmissions = pd.read_csv("dataset/annual-co-emissions-by-region.csv")
annual_co_emmissions["GlobalCO2Mean"] = annual_co_emmissions.groupby("Year")["Annual CO₂ emissions"].transform("mean").round()
annual_co_emmissions.head()
```

### Pivot the CO₂ Emissions Data to Wide Format
Here we pivot the annual CO₂ emissions data so that each country's ("Entity") emissions become
a separate column, with "Year" as the index. This allows us to merge country-specific emissions
data with our merged climate dataset.

```python
co2_wide = annual_co_emmissions.pivot(index='Year', columns='Entity', values='Annual CO₂ emissions')
co2_wide.head()
```

### Merge All Datasets
We merge the climate dataset (which now includes temperature and sea ice data) with the pivoted
CO₂ emissions data on "Year". The merge is done as a left join so that all years in `merged_tempice`
are retained, and corresponding country-level emissions are added.

```python
merged_all = pd.merge(merged_tempice, co2_wide, on='Year', how='left')
merged_all.head()
```

### Add the Global CO₂ Mean to the Final Merge
Since the pivot does not include the "GlobalCO2Mean" column, we create a separate DataFrame for it, remove duplicates, and then merge it with the final dataset.

```python
global_co2 = annual_co_emmissions[['Year', 'GlobalCO2Mean']].drop_duplicates()
merged_all = pd.merge(merged_all, global_co2, on='Year', how='left')
merged_all.head()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Compute standardized anomalies (z-scores) for each variable:
merged_all["Arctic_z"] = (merged_all["64N-90N"] - merged_all["64N-90N"].mean()) / merged_all["64N-90N"].std()
merged_all["SeaIce_z"] = (merged_all["SeaIceMean"] - merged_all["SeaIceMean"].mean()) / merged_all["SeaIceMean"].std()
merged_all["GlobCO2Mean_z"] = (merged_all["GlobalCO2Mean"] - merged_all["GlobalCO2Mean"].mean()) / merged_all["GlobalCO2Mean"].std()


# Invert the Sea Ice anomaly (so that a decrease appears as a positive anomaly)
merged_all["SeaIce_z_inv"] = -merged_all["SeaIce_z"]

# Create the figure
fig = go.Figure()

# Arctic Temperature anomaly (red, always visible)
fig.add_trace(go.Scatter(
    x=merged_all["Year"],
    y=merged_all["Arctic_z"],
    mode="lines+markers",
    name="Arctic Temperature (z-score)",
    line=dict(color="red", width=2)
))

# Inverted Sea Ice anomaly (blue) – default visible
fig.add_trace(go.Scatter(
    x=merged_all["Year"],
    y=merged_all["SeaIce_z_inv"],
    mode="lines+markers",
    name="Sea Ice Extent (inverted, z-score)",
    line=dict(color="blue", width=2),
    visible=False
))

# Non-inverted Sea Ice anomaly (blue) – default hidden
fig.add_trace(go.Scatter(
    x=merged_all["Year"],
    y=merged_all["SeaIce_z"],
    mode="lines+markers",
    name="Sea Ice Extent (z-score)",
    line=dict(color="blue", width=2),
    visible=True
))

fig.add_trace(go.Scatter(
    x=merged_all["Year"],
    y=merged_all["GlobCO2Mean_z"],
    mode="lines+markers",
    name="Global CO2 Mean (z-score)",
    line=dict(color="orange", width=2),
    visible=True
))



# Create update menu buttons to toggle between inverted and non-inverted Sea Ice series
updatemenus = [
    {
        "buttons": [
            {
                "label": "Normal",
                "method": "update",
                "args": [{"visible": [True, False, True]},
                         {"title": "Annual Anomalies: Arctic Temperature vs. Sea Ice Extent"}]
            },
                        {
                "label": "Inverted",
                "method": "update",
                "args": [{"visible": [True, True, False]},
                         {"title": "Annual Anomalies: Arctic Temperature vs. Inverted Sea Ice Extent"}]
            },
        ],
        "direction": "down",
        "showactive": True,
        "x": 0.0,
        "xanchor": "left",
        "y": 1.15,
        "yanchor": "top"
    }
]

fig.update_layout(
    updatemenus=updatemenus,
    title="Annual Anomalies: Arctic Temperature vs. Inverted Sea Ice Extent",
    xaxis_title="Year",
    yaxis_title="Standardized Anomaly (z-score)",
    template="plotly_white",
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig.show()
```
